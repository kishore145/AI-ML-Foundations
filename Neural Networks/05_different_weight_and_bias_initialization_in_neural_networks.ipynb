```
# Import depdendencies
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.initializers import Zeros, RandomNormal, RandomUniform, glorot_normal, glorot_uniform
```

```
# Set number of neurons
n_input = 784
n_dense = 256
```

```
# Initializers allow you to pre-specify an initialization strategy, 
# encoded in the Initializer object, without knowing the shape and dtype of the variable being initialized

# Usually bias vector is initialized with zeros. 
b_init = Zeros()
```

```
# Weight initializers created as a list to visualize how each one works
w_inits = [Zeros(), RandomNormal(stddev=1.0), RandomUniform( ), glorot_normal(), glorot_uniform()]
```

```
# Different activation functions to use
a_functions = ['relu', 'sigmoid', 'tanh']
```

```
# Design a shallow neural network
def create_model(w_init, b_init, a_function):
  model = Sequential()
  model.add(Dense( n_dense, input_dim = n_input, 
                  kernel_initializer = w_init,
                  bias_initializer = b_init))
  model.add(Activation(a_function))
  return model
```

```
# Generate a random input sample
x = np.random.random((1,n_input))
x.shape
```

```
# Forward propogate the model using different weights and activation functions
# for labelling graphs
weights = ['Zeros', 'RandomNormal', 'RandomUniform', 'glorot_normal', 'glorot_uniform'] 


# Looping through different activation functions and weight initializers
for a_function in a_functions:
  fig, axs = plt.subplots(1, 5, constrained_layout=True, figsize=(25,4.5))
  fig.suptitle(f'Actication Function: {a_function}')
  i = 0 # Reset i for next activation function
  for w_init in w_inits:
    # Create a model and do one round of forwardd propogation
    model = create_model(w_init, b_init, a_function) 
    a = model.predict(x) # a is the o/p for first layer of neurons
    
    # Plot the results in a histogram
    axs[i].set_title(f'{weights[i]}', loc='center' )
    axs[i].hist(np.transpose(a))
    i += 1
  
  


```
